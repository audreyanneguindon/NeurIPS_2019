## Results

```python
import pandas as pd
import numpy as np
from scipy.stats import gmean, gstd

policy_code = {
    'gcnn:mean_convolution': 'gcnn-mean',
    'gcnn:no_prenorm': 'gcnn-sum',
    'gcnn:baseline': 'gcnn',
    'internal:relpscost': 'rpb',
    'internal:relpscost-cuts': 'rpb-cuts',
    'internal:relpscost-nocuts': 'rpb-nocuts',
    'internal:fullstrong': 'fsb',
    'internal:vanillafullstrong': 'v-fsb',
}

policies = [
    'internal:fullstrong',
    'internal:vanillafullstrong',
    'internal:relpscost',
    'gcnn:baseline',
    'gcnn:mean_convolution',
    'gcnn:no_prenorm'
]
```

### Policy Evaluation 

```python
# cauctions
results = pd.concat([
    pd.read_csv("results/cauctions_20191128-033332.csv"),
], axis=0)

print(results.shape[0] / 300)

results = results[results.policy.isin(policies)]
results.to_csv("results/tmp.csv", index=False)

results['solved'] = (results.status == "optimal")

has_solved = results[['type', 'policy', 'solved']].groupby(['type', 'policy']).agg({'solved': ['max']})
has_solved.columns = has_solved.columns.map('_'.join)
has_solved = has_solved.rename(columns={'solved_max': 'has_solved'})
results = pd.merge(results, has_solved, on=['type', 'policy'])

# instances which all methods (except methods which solved 0 instances of that type) have solved
all_solved = results[results.has_solved][['instance', 'seed', 'solved']].groupby(['instance', 'seed']).agg({'solved': ['min']})
all_solved.columns = all_solved.columns.map('_'.join)
all_solved = all_solved.rename(columns={'solved_min': 'all_solved'})
results = pd.merge(results, all_solved, on=['instance', 'seed'])
```

```python
result_stats = {}
time_shift = 1
nnodes_shift = 1
for type_ in results['type'].unique():
    for policy in results['policy'].unique():
        tp_res = results[(results.policy == policy) & (results.type == type_)]
        tp_res_nc = tp_res[tp_res.all_solved & tp_res.has_solved]
        result_stats[(policy, type_)] = {
            'time': gmean(tp_res.stime + time_shift) - time_shift, 
#             'time_std': np.std(tp_res.stime), 
            'time_std': 100 * np.mean([np.std(tp_res[tp_res.instance == instance].stime) / np.mean(tp_res[tp_res.instance == instance].stime) for instance in tp_res['instance'].unique()]),
            'nnodes': gmean(tp_res_nc.nnodes + nnodes_shift) - nnodes_shift,
#             'nnodes_std': np.std(tp_res_nc.nnodes),
            'nnodes_std': 100 * np.mean([np.std(tp_res_nc[tp_res_nc.instance == instance].nnodes) / np.mean(tp_res_nc[tp_res_nc.instance == instance].nnodes) for instance in tp_res_nc['instance'].unique()]),
            'nnodes_fair': gmean(tp_res_nc.nnodes + 2*(tp_res_nc.ndomchgs + tp_res_nc.ncutoffs) + nnodes_shift) - nnodes_shift,
            'nnodes_fair_std': np.std(tp_res_nc.nnodes + 2*(tp_res_nc.ndomchgs + tp_res_nc.ncutoffs)),
            'completed': sum(tp_res.solved),
            'winner': 0,
        }
    for instance in results[results.type == type_].instance.unique():
        for seed in results['seed'].unique():
            seed_results = results[(results.type == type_) & (results.instance == instance) & (results.seed == seed) & (results.solved == 1)]
            if seed_results.shape[0] > 0:
                winner = seed_results.loc[seed_results.stime.idxmin()].policy
                result_stats[(winner, type_)]['winner'] += 1
```

```python
for type_ in ('small', 'medium', 'big'):
    print(type_)
    for policy in policies:
        if policy in results['policy'].unique():
            stats = result_stats[(policy, type_)]
            print(f"    {policy_code[policy]: <10}:",
                  f"  time {stats['time']:7.2f} +- {stats['time_std']:2.1f} %"
                  f"  wins {stats['winner']:3d}/{stats['completed']:3d}"
                  f"  nodes {stats['nnodes']:5.0f} +- {stats['nnodes_std']:4.1f} %"
                  f"  ({stats['nnodes_fair']:5.0f})"
            )
```

### Accuracy

```python
results = pd.concat([
    pd.read_csv("results/cauctions_test_20191128-025805.csv")
])

results = results.groupby(['policy']).agg({f"acc@{i}": ['mean', 'std'] for i in (1, 3, 5, 10)}).reset_index()
results.columns = results.columns.map('_'.join)

for policy in [
    'gcnn:baseline'
]:
    print(f"{policy_code[policy]: <3}")
    for i in (1, 5, 10):
        acc = float(results[results.policy_ == policy][f'acc@{i}_mean'])
        std = float(results[results.policy_ == policy][f'acc@{i}_std'])
        print(f"        & {100*acc:0.1f} & {100*std:0.1f}")

        print("\\\\")
```
